```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("noob786/geoai-challenge-for-cropland-mapping-dry-dataset")

print("Path to dataset files:", path)

```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from shapely.geometry import Point
from sklearn.neighbors import KDTree

# Define the dataset path
dataset_path = "/root/.cache/kagglehub/datasets/noob786/geoai-challenge-for-cropland-mapping-dry-dataset/versions/1"

# Load Sentinel-1 and 2 and drop date
s1 = pd.read_csv(f"{dataset_path}/Sentinel1.csv").drop(columns=["date"])
s2 = pd.read_csv(f"{dataset_path}/Sentinel2.csv").drop(columns=["date"])

# Load shapefiles and extract labeled geodata
fergana_gdf = gpd.read_file(f"{dataset_path}/Train/Fergana_training_samples.shp")
orenburg_gdf = gpd.read_file(f"{dataset_path}/Train/Orenburg_training_samples.shp")
train_gdf = pd.concat([fergana_gdf, orenburg_gdf])
train_gdf = train_gdf[['Cropland', 'geometry']]
train_gdf["lon"] = train_gdf.geometry.x
train_gdf["lat"] = train_gdf.geometry.y

# Match Sentinel2 ID points to nearest labeled point
tree = KDTree(train_gdf[["lat", "lon"]].values)
s2_points = s2.groupby("ID")[["translated_lat", "translated_lon"]].mean().reset_index()
dist, idx = tree.query(s2_points[["translated_lat", "translated_lon"]].values, k=1)
s2_points["label"] = train_gdf.iloc[idx.flatten()].Cropland.values
s2_labels = s2_points[["ID", "label"]]

# Aggregate function (excluding ID)
def aggregate_features(df, id_col="ID"):
    return df.drop(columns=["translated_lat", "translated_lon"], errors='ignore') \
             .groupby(id_col).agg(['mean', 'std', 'min', 'max']).reset_index()

# Aggregate train features
s1_feats = aggregate_features(s1[["ID", "VH", "VV"]])
s2_feats = aggregate_features(s2[["ID", "B2", "B3", "B4", "B5", "B6", "B7", "B8", "B8A", "B11", "B12"]])

# Flatten multi-index columns
s1_feats.columns = ['_'.join(col).strip("_") for col in s1_feats.columns.values]
s2_feats.columns = ['_'.join(col).strip("_") for col in s2_feats.columns.values]

# Merge training features and labels
train_df = s2_feats.merge(s1_feats, on="ID", how="outer").merge(s2_labels, on="ID", how="inner")
train_df = train_df.dropna()

# Train-test split
X = train_df.drop(columns=["ID", "label"])
y = train_df["label"]
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train XGBoost
model = xgb.XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.05, tree_method="hist", random_state=42)
model.fit(X_train, y_train)

# Validate
val_preds = model.predict(X_val)
acc = accuracy_score(y_val, val_preds)
print(f"âœ… Validation Accuracy: {acc:.4f}")

# Prepare test set
test_meta = pd.read_csv(f"{dataset_path}/Test.csv")
test_ids = test_meta["ID"].unique()
s1_test = s1[s1["ID"].isin(test_ids)]
s2_test = s2[s2["ID"].isin(test_ids)]

# Drop date + aggregate test features
s1_test_feats = aggregate_features(s1_test[["ID", "VH", "VV"]])
s2_test_feats = aggregate_features(s2_test[["ID", "B2", "B3", "B4", "B5", "B6", "B7", "B8", "B8A", "B11", "B12"]])
s1_test_feats.columns = ['_'.join(col).strip("_") for col in s1_test_feats.columns.values]
s2_test_feats.columns = ['_'.join(col).strip("_") for col in s2_test_feats.columns.values]

# Merge test features
test_df = s2_test_feats.merge(s1_test_feats, on="ID", how="outer").fillna(0)
X_test = test_df.drop(columns=["ID"])
test_preds = model.predict(X_test)

# Final submission
submission = pd.read_csv(f"{dataset_path}/SampleSubmission.csv")
submission = submission.drop(columns=["Cropland"])
submission = submission.merge(test_df[["ID"]], on="ID", how="right")
submission["Cropland"] = test_preds.astype(int)
submission.to_csv("submission.csv", index=False)
print("ðŸ“¤ Saved submission.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Display descriptive statistics to get an overview of the data distribution
display(X_very_large.describe())

# Create box plots for a subset of features to visualize potential outliers
# Due to the number of features, we'll plot a few to demonstrate.
# You can modify the list of features to explore others.
features_to_plot = ['B2_mean', 'B3_mean', 'B4_mean', 'NDVI_mean', 'EVI_mean', 'VH_mean', 'VV_mean']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot):
    plt.subplot(2, 4, i + 1)
    sns.boxplot(x=X_very_large[feature])
    plt.title(feature)
plt.tight_layout()
plt.show()
```

**Reasoning**:
Remove outliers from the features using the Interquartile Range (IQR) method.

```python
# Calculate Q1, Q3, and IQR for each feature
Q1 = X_very_large.quantile(0.25)
Q3 = X_very_large.quantile(0.75)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a boolean mask to identify outliers
outlier_mask = (X_very_large < lower_bound) | (X_very_large > upper_bound)

# Get the indices of rows that have at least one outlier
outlier_indices = X_very_large.index[outlier_mask.any(axis=1)]

# Remove rows with outliers from the feature and target dataframes
X_very_large_cleaned = X_very_large.drop(outlier_indices)
y_very_large_cleaned = y_very_large.drop(outlier_indices)

print(f"Original number of samples: {len(X_very_large)}")
print(f"Number of samples after outlier removal: {len(X_very_large_cleaned)}")

# Display the head of the cleaned feature dataframe
display(X_very_large_cleaned.head())
```

## Hyperparameter tuning

### Subtask:
Perform hyperparameter tuning on the trained LightGBM model to optimize its performance.

Import necessary libraries for hyperparameter tuning and define the parameter grid for LightGBM.

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 6, 9]
}
```
